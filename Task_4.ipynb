```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/modified_road_accident_dataset.csv')
```

```python
df
```

```python
df.columns
```

```python
#Accident count by Road conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Road Condition', order=df['Road Condition'].value_counts().index)
plt.xlabel('Number of accidents')
plt.ylabel('Road Condition')
plt.title('Accident count by Road Conditions')
plt.show()
```

```python
# Distribution of accident by Weather conditions
weather_counts = df['Weather Conditions'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('magma', len(weather_counts)))
plt.title('Distribution of accidents by Weather Conditions')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#Distribution of Driver Alcohol level
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Driver Alcohol Level', kde=True)
plt.xlabel('Driver Alcohol Level')
plt.ylabel('Frequency')
plt.title('Distribution of Driver Alcohol Level')
plt.show()
```

```python
# Accident count by Time of day
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Time of Day', order=df['Time of Day'].value_counts().index, palette='viridis')
plt.xlabel('Number of accidents')
plt.ylabel('Time of Day')
plt.title('Accident count by Time of Day')
plt.show()
```

```python
#Distribution of NUmber of vehicals involved by region
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Region', y='Number of Vehicles Involved', palette='viridis')
plt.xlabel('Region')
plt.ylabel('Number of Vehicles Involved')
plt.title('Distribution of Number of Vehicles Involved by Region')
```

```python
#Top 5 accident hotspots
top_5_regions = df['Region'].value_counts().nlargest(5).index
df_top_5_regions = df[df['Region'].isin(top_5_regions)]

plt.figure(figsize=(10, 6))
sns.countplot(data=df_top_5_regions, x='Region', order=top_5_regions, palette='viridis')
plt.xlabel('Region')
plt.ylabel('Number of accidents')
plt.title('Top 5 Regions by Accident Count')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```
